```python
import numpy as np
```

```python
a = np.array([1,2,3])
```

```python
b = np.array([3,2,1])
```

```python
dot = 0
for e, f in zip(a,b):
    dot += e*f
```

```python
dot
```

```python
a@b
```

```python
amag = np.sqrt(a@a)
```

```python
bmag = np.sqrt(b@b)
```

```python
cos = (a@b)/(bmag*amag)
cos
```
