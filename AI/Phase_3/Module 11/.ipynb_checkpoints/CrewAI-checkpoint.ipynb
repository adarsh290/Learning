```python
from typing import List, Dict, Any
```

```python
# --- Step 1: Define Structural Blueprint Classes (Emulating CrewAI Core) ---

class CrewAgent:
    """Represents a clean, self-contained worker profile configuration."""
    def __init__(self, role: str, goal: str, backstory: str):
        self.role = role
        self.goal = goal
        self.backstory = backstory

    def execute(self, task_description: str, context_input: str = "") -> str:
        """Simulates execution within the specific role and backstory prompt boundaries."""
        print(f"\n  [AGENT START] {self.role} is initializing task assignment...")
        print(f"  [Context Anchor] Backstory: {self.backstory}")
        
        # Simulated intelligent output processing based on role definitions
        if "researcher" in self.role.lower():
            return f"RESEARCH_OUTPUT: Located 3 major 2026 academic publications indicating that matching hybrid-perovskite crystals with customized silicon layers pushes tandem solar cell efficiency targets straight to 31.5%."
        elif "writer" in self.role.lower():
            cleaned_context = context_input.replace("RESEARCH_OUTPUT:", "").strip()
            return f"TECHNICAL_ARTICLE_DRAFT:\n  Title: Breaking the Solar Ceiling in 2026\n  Body: Recent studies reveal hybrid-perovskite/silicon tandem modules have achieved a landmark 31.5% efficiency rate by optimization of crystal layer matching."
            
        return "Generic task processing complete."

class CrewTask:
    """Represents an isolated, measurable objective work item."""
    def __init__(self, description: str, expected_output: str, assigned_agent: CrewAgent):
        self.description = description
        self.expected_output = expected_output
        self.agent = assigned_agent
        self.output_result = ""

class CrewOrchestrator:
    """Manages the execution flow and context handoffs across the team layout."""
    def __init__(self, agents: List[CrewAgent], tasks: List[CrewTask], process_mode: str = "sequential"):
        self.agents = agents
        self.tasks = tasks
        self.process_mode = process_mode
        print(f"[CREW] Team assembled. Execution Policy set to: '{self.process_mode.upper()}'")

    def kickoff(self) -> str:
        current_context = ""
        
        # Simulating automated sequential pipeline execution
        for index, task in enumerate(self.tasks, start=1):
            print(f"\n=== Executing Team Task {index}/{len(self.tasks)} ===")
            print(f"Task Directive: {task.description}")
            
            # Fire execution block on the assigned agent node, passing past context along seamlessly
            task.output_result = task.agent.execute(task.description, context_input=current_context)
            
            # The output of this task immediately becomes the structural context for the next agent
            current_context = task.output_result
            print(f"Task Complete! Captured Output Length: {len(task.output_result)} characters.")
            
        return current_context
```

```python
# --- Step 2: Assemble the Crew and Kick Off Production Workflow ---

# 1. Instantiate the individual specialized agent nodes
materials_scout = CrewAgent(
    role="Senior Solar Materials Researcher",
    goal="Extract the latest 2026 laboratory efficiency metrics for advanced solar cells.",
    backstory="You are an analytical data scientist. You read academic literature and isolate raw metrics with absolute precision."
)

tech_copywriter = CrewAgent(
    role="Lead Technology Communications Writer",
    goal="Translate dense experimental academic metrics into highly engaging public articles.",
    backstory="You are a veteran scientific journalist. You take raw mathematical discoveries and synthesize them into clear, compelling narratives."
)

# 2. Map out the explicit task sequences
task_1 = CrewTask(
    description="Scan the 2026 materials informatics database for breakthrough solar cell bandgap efficiency numbers.",
    expected_output="A single string containing precise material components and validated efficiency percentages.",
    assigned_agent=materials_scout
)

task_2 = CrewTask(
    description="Review the raw data outputted by the researcher and draft a compelling technical announcement article.",
    expected_output="A structured markdown article containing a title and synthesized summary paragraphs.",
    assigned_agent=tech_copywriter
)

# 3. Instantiate the master Crew orchestrator and start execution
solar_project_crew = CrewOrchestrator(
    agents=[materials_scout, tech_copywriter],
    tasks=[task_1, task_2],
    process_mode="sequential"
)
```

```python
# Run the pipeline
final_crew_deliverable = solar_project_crew.kickoff()

print("\n" + "="*60)
print("FINAL CREW DELIVERABLE EXPORTED TO CLIENT:")
print("="*60)
print(final_crew_deliverable)
```
