```python
# --- Step 1: Define Available Environment Tools ---
def get_rack_temperature(rack_id: str) -> str:
    """A real-world execution tool checking live server environments."""
    if rack_id == "rack_4":
        return "42°C (CRITICAL OVERHEAT ALERT)"
    return "24°C (Normal)"

def fetch_on_call_engineer(zone: str) -> str:
    """A database lookup tool retrieving team assignments."""
    return "Aryan Yadav (Email: aryan.yadav@iitropar.ac.in)"

def dispatch_email_alert(recipient: str, message: str) -> str:
    """Simulates an external API integration call sending an alert notification."""
    return f"SUCCESS: Email dispatch confirmed to {recipient}."
```

```python
# --- Step 2: The Core ReAct Orchestration Engine Loop ---
class FirstPrinciplesReActAgent:
    def __init__(self):
        print("[AGENT ENGINE] Initializing autonomous loop architecture...")

    def run_agent_loop(self, user_objective: str):
        # Initialize the baseline execution history memory envelope
        context_history = f"""
SYSTEM RULES:
You are an autonomous operations agent. You solve tasks by alternating between Thought, Action, and Observation.
Available Tools:
- get_rack_temperature[rack_id]: Returns current temperature string.
- fetch_on_call_engineer[zone]: Returns name and email details.
- dispatch_email_alert[recipient, message]: Sends an email notification.

When you have completed your objective, your final line must read exactly: "FINAL_ANSWER: [Your summary message]"

OBJECTIVE: {user_objective}
"""
        # We manually simulate the multi-turn execution step-by-step
        # Turn 1: Initial Analysis
        print("\n--- AGENT TURN 1 ---")
        thought_1 = "Thought: I need to determine if server rack 4 is overheating. I must use get_rack_temperature tool first."
        action_1 = "Action: get_rack_temperature[rack_4]"
        print(thought_1)
        print(action_1)
        
        # System executes the real code tool based on the Action instruction string
        obs_1_text = get_rack_temperature("rack_4")
        observation_1 = f"Observation: {obs_1_text}"
        print(observation_1)
        
        # Append Turn 1 variables to our memory track
        context_history += f"\n{thought_1}\n{action_1}\n{observation_1}\n"
        
        # Turn 2: Processing the observation data and planning step 2
        print("\n--- AGENT TURN 2 ---")
        thought_2 = "Thought: The temperature is 42°C, which is a critical overheat. I must now find out who the on-call engineer is for this zone to alert them."
        action_2 = "Action: fetch_on_call_engineer[server_room_east]"
        print(thought_2)
        print(action_2)
        
        obs_2_text = fetch_on_call_engineer("server_room_east")
        observation_2 = f"Observation: {obs_2_text}"
        print(observation_2)
        
        context_history += f"\n{thought_2}\n{action_2}\n{observation_2}\n"
        
        # Turn 3: Executing the final corrective action step
        print("\n--- AGENT TURN 3 ---")
        thought_3 = "Thought: I have the engineer's email address (aryan.yadav@iitropar.ac.in). I will now dispatch the critical overheat notification alert."
        action_3 = "Action: dispatch_email_alert[aryan.yadav@iitropar.ac.in, Emergency: Rack 4 is critical!]"
        print(thought_3)
        print(action_3)
        
        obs_3_text = dispatch_email_alert("aryan.yadav@iitropar.ac.in", "Emergency: Rack 4 is critical!")
        observation_3 = f"Observation: {obs_3_text}"
        print(observation_3)
        
        context_history += f"\n{thought_3}\n{action_3}\n{observation_3}\n"
        
        # Turn 4: Final verification and closure
        print("\n--- AGENT TURN 4 ---")
        final_thought = "Thought: The alert has been sent successfully. The task objective is completely fulfilled."
        final_verdict = "FINAL_ANSWER: Overheat condition identified on rack 4 (42°C). Incident handled and email notification successfully routed to on-call infrastructure lead Aryan Yadav."
        print(final_thought)
        print(final_verdict)
```

```python
# Execute the manual ReAct logic trace
agent = FirstPrinciplesReActAgent()
agent.run_agent_loop("Check rack 4 metrics and notify the responsible technician if it is burning up.")
```
