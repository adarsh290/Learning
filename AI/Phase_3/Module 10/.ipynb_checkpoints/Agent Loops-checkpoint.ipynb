```python
import json
from typing import Dict, Any, List
```

```python
# --- Simulated Tool Environment ---
def check_bandgap_database(material_id: str) -> str:
    """Simulated database tracking solar material characteristics."""
    if "perovskite" in material_id.lower():
        return "Bandgap Energy: 1.55 eV (Optimal for single-junction solar conversion)"
    return "Error: Material compound matching index not located."
```

```python
# --- The Autonomous Agent Loop Core Engine ---
class AutonomousAgentLoopEngine:
    def __init__(self, max_allowed_turns: int = 4):
        self.max_turns = max_allowed_turns
        print(f"[ENGINE] Autonomous Agent Container active. Safety ceiling: {self.max_turns} maximum loops.")

    def simulate_llm_cognitive_step(self, turn: int, history: List[str]) -> Dict[str, Any]:
        """
        Simulates the model's dual-track processing. 
        It returns either a tool request OR a termination signal.
        """
        # Turn 1: Model decides it needs to gather data
        if turn == 1:
            return {
                "thought": "I need to find the specific bandgap value of Perovskite solar compounds. I will use the database tool.",
                "action": "CALL_TOOL",
                "tool_name": "check_bandgap_database",
                "tool_args": {"material_id": "hybrid-perovskite"}
            }
        
        # Turn 2: Model reads the observation from turn 1, evaluates progress, and completes the task
        if turn == 2:
            return {
                "thought": "I have retrieved the precise bandgap data (1.55 eV) from the reference system. I can now fulfill the user's objective completely.",
                "action": "HALT_AND_RETURN",
                "final_verdict": "Perovskite solar materials exhibit an optimal bandgap energy of 1.55 eV for single-junction photovoltaics."
            }
            
        return {"action": "HALT_AND_RETURN", "final_verdict": "Timeout unexpected trace."}

    def execute(self, user_goal: str):
        print(f"Goal Ingested: '{user_goal}'\n")
        
        execution_history: List[str] = []
        current_turn = 1
        is_running = True
        
        # The Core Agent Runtime Loop
        while is_running:
            print(f"--- LOOP TURN {current_turn} of {self.max_turns} ---")
            
            # 1. Check defensive circuit breaker boundary
            if current_turn > self.max_turns:
                print("[CRITICAL SHUTDOWN] Agent hit max iteration cap! Terminating to safeguard resources.")
                print("Returning partial diagnostic telemetry back to user.")
                break
                
            # 2. Execute internal evaluation and generation step
            llm_output = self.simulate_llm_cognitive_step(current_turn, execution_history)
            
            print(f"  Reasoning: {llm_output['thought']}")
            
            # 3. Route actions based on the structural token command returned
            command = llm_output["action"]
            
            if command == "CALL_TOOL":
                tool_target = llm_output["tool_name"]
                args = llm_output["tool_args"]
                
                print(f"  Command: [Invoke Tool] → Execute '{tool_target}' with payload: {args}")
                
                # Execute tool tracking integration
                observation_result = check_bandgap_database(**args)
                print(f"  Observation Captured: {observation_result}")
                
                # Update history log states
                execution_history.append(llm_output["thought"])
                execution_history.append(observation_result)
                
            elif command == "HALT_AND_RETURN":
                # 4. Success terminal edge hit!
                print("\n[STOP SIGNAL DETECTED] Agent has declared goal fulfillment.")
                print("Closing tool channels and breaking runtime loop.")
                print("="*60)
                print("FINAL SYSTEM DISPATCH VERDICT:")
                print("="*60)
                print(llm_output["final_verdict"])
                is_running = False
                
            # Increment loop counter state securely
            current_turn += 1
```

```python
# Execute the runner instance
agent_runtime = AutonomousAgentLoopEngine(max_allowed_turns=4)
agent_runtime.execute("Find the bandgap metric for perovskite solar components.")
```
