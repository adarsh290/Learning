```python
import math
from typing import List
```

```python
def dot_product(vector_a: List[float], vector_b: List[float]) -> float:
    """Computes the sum of products of corresponding elements."""
    return sum(a * b for a, b in zip(vector_a, vector_b))

def magnitude(vector: List[float]) -> float:
    """Computes the geometric length (norm) of the vector from the origin."""
    return math.sqrt(sum(x**2 for x in vector))

def calculate_cosine_similarity(vector_a: List[float], vector_b: List[float]) -> float:
    """Calculates the exact angular similarity score between two numeric vectors."""
    mag_a = magnitude(vector_a)
    mag_b = magnitude(vector_b)
    
    if mag_a == 0 or mag_b == 0:
        return 0.0 # Guard against division by zero errors
        
    return dot_product(vector_a, vector_b) / (mag_a * mag_b)
```

```python
# --- Mock 3-Dimensional Embedding Registry ---
# Imagine a small model maps text to 3 specific latent dimensions:
# Dimension 1: Technology Focus | Dimension 2: Biological Focus | Dimension 3: Financial Focus

mock_embeddings = {
    "quantum_computing": [0.95, 0.05, 0.40],
    "artificial_intelligence": [0.90, 0.12, 0.35],
    "cellular_mitosis": [0.02, 0.98, 0.01],
    "investment_portfolio": [0.30, 0.01, 0.92]
}
```

```python
# --- Compare Concepts ---

# 1. Tech vs Tech
sim_tech_vs_tech = calculate_cosine_similarity(
    mock_embeddings["quantum_computing"], 
    mock_embeddings["artificial_intelligence"]
)

# 2. Tech vs Biology
sim_tech_vs_bio = calculate_cosine_similarity(
    mock_embeddings["quantum_computing"], 
    mock_embeddings["cellular_mitosis"]
)

# 3. Tech vs Finance
sim_tech_vs_finance = calculate_cosine_similarity(
    mock_embeddings["quantum_computing"], 
    mock_embeddings["investment_portfolio"]
)
```
