```python
from typing import List

def sliding_window_chunker(text: str, chunk_size: int, chunk_overlap: int) -> List[str]:
    """
    Slices a continuous text string into small, overlapping blocks 
    using character count boundaries.
    """
    # Guard check: Overlap cannot be larger than or equal to the total chunk capacity
    if chunk_overlap >= chunk_size:
        raise ValueError("Overlap size must be strictly smaller than the total chunk size.")
        
    chunks = []
    start_index = 0
    text_length = len(text)
    
    print(f"[CHUNKER] Processing raw text source containing {text_length} characters...")
    print(f"[CHUNKER] Configuration -> Size: {chunk_size} chars | Overlap: {chunk_overlap} chars\n")
    
    while start_index < text_length:
        # Calculate the end boundary index of the current window segment
        end_index = start_index + chunk_size
        
        # Extract the slice from the string
        chunk_slice = text[start_index:end_index]
        chunks.append(chunk_slice)
        
        # PITFALL WARNING: If we advanced by chunk_size, we would lose contextual continuity.
        # Instead, we advance by (chunk_size - chunk_overlap) to create the buffer zone.
        start_index += (chunk_size - chunk_overlap)
        
    return chunks
```

```python
# --- Test Document Data ---
document_payload = (
    "The Alpha-9 engine variant utilizes liquid methane propellant. "
    "This choice optimizes specific impulse efficiency at sea level. "
    "Conversely, the Beta-3 system relies on liquid hydrogen for deep space operations. "
    "Hydrogen fuel systems require specialized cryogenic insulation tanks to mitigate boiling."
)
```

```python
# Execute the chunking operation
generated_chunks = sliding_window_chunker(
    text=document_payload, 
    chunk_size=80, 
    chunk_overlap=25
)
```

```python
# Inspect the structured results
print("="*60)
print("INSPECTING CHUNK GENERATION RUN OUTPUT:")
print("="*60)
for index, chunk in enumerate(generated_chunks, start=1):
    print(f"Chunk #{index} (Length: {len(chunk)} chars):")
    print(f"  → \"{chunk}\"")
    print("-" * 40)
```
