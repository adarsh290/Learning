```python
from typing import List, Dict, Any
```

```python
# --- Step 1: Simulated Database Setup (Our Private Knowledge Base) ---
RAW_KNOWLEDGE_BASE = [
    {
        "id": "doc_1",
        "text": "The PH102 Lab Team successfully tested the Plasma Singing Arc. It generates an audible plasma audio signal using a solid-state flyback transformer and a ZVS (Zero Voltage Switching) driver core."
    },
    {
        "id": "doc_2",
        "text": "For the Plasma Singing Arc, the primary switching transistors selected were IRFP260N power MOSFETs. These components require heavy aluminum heatsinks and active fan cooling to prevent thermal runaway at 24V."
    },
    {
        "id": "doc_3",
        "text": "The CampusGrid project is an intranet-based local network application designed to handle high-speed digital game distributions exclusively inside college hostel networks using chunked file streaming profiles."
    }
]
```

```python
# --- Step 2: The Core RAG Engine Implementation ---
class FullRAGPipeline:
    def __init__(self, database_source: List[Dict[str, str]]):
        self.db = database_source
        print("[RAG Engine] System online. Knowledge Base indexed successfully.")

    def retrieve_relevant_context(self, user_query: str) -> str:
        """
        Simulates the entire Retrieval + Reranking loop.
        Scans our internal database for the best matching context blocks.
        """
        print(f"[Retrieval] Scanning index for keywords and concepts matching: '{user_query}'")
        
        query_words = set(user_query.lower().replace("?", "").split())
        matched_chunks = []

        # Simple semantic/keyword similarity matching simulation
        for record in self.db:
            record_text_lower = record["text"].lower()
            # Calculate how many query words intersect with this specific database record
            overlap_score = sum(1 for word in query_words if word in record_text_lower)
            
            if overlap_score > 0:
                matched_chunks.append((record["text"], overlap_score))
        
        # Sort matched text records by their calculated score relevance descending
        matched_chunks.sort(key=lambda x: x[1], reverse=True)
        
        # Extract the text from the single best matching block
        if matched_chunks:
            best_context = matched_chunks[0][0]
            print(f"[Retrieval] Found optimum context match. Extracted {len(best_context)} characters.")
            return best_context
        
        print("[Retrieval WARNING] No relevant documents found in local database indexing.")
        return "No specific reference documentation located in corporate knowledge pools."

    def augment_prompt(self, query: str, context: str) -> str:
        """
        The Augmentation Step.
        Combines the system guardrails, retrieved documents, and user query into one prompt.
        """
        print("[Augmentation] Injecting retrieved facts into the LLM system prompt envelope...")
        
        # PITFALL WARNING: We must explicitly instruct the model to use ONLY this context 
        # to prevent it from hallucinating or falling back on its general training memory.
        augmented_template = f"""
SYSTEM INSTRUCTIONS:
You are a precise engineering assistant. Answer the user's question using ONLY the verified facts provided below. 
If the context does not contain the answer, state explicitly: "Information not found in local files."
Do not make assumptions or invent details outside the provided source material.

VERIFIED SOURCE CONTEXT:
\"\"\"
{context}
\"\"\"

USER QUESTION: 
{query}

CONCISE REASONED ANSWER:"""
        return augmented_template

    def generate_response(self, complete_prompt: str) -> str:
        """The Generation Step. Simulates the final LLM text generation."""
        print("[Generation] Dispatching augmented context block to LLM...")
        
        # Simulated LLM internal logic responding strictly to the prompt text structure
        if "IRFP260N" in complete_prompt:
            return "The Plasma Singing Arc uses IRFP260N power MOSFETs as its primary switching transistors. To keep them safe from thermal runaway at 24V, you must equip them with heavy aluminum heatsinks and active fan cooling."
        elif "CampusGrid" in complete_prompt:
            return "CampusGrid is an intranet game distribution platform that uses chunked file streaming."
        else:
            return "Information not found in local files."

    def execute_rag_flow(self, user_question: str) -> str:
        """Executes the complete, end-to-end RAG automated runtime loop."""
        print(f"\n=== STARTING FULL RAG TRANSACTION ===")
        
        # 1. Fetch relevant knowledge
        context = self.retrieve_relevant_context(user_question)
        
        # 2. Build the structural context envelope prompt
        final_prompt = self.augment_prompt(user_question, context)
        
        # 3. Get the final structured textual generation from the LLM core
        answer = self.generate_response(final_prompt)
        
        print(f"=== RAG TRANSACTION COMPLETED SUCCESSFULLY ===\n")
        return answer
```

```python
# --- Step 3: Run the Complete Pipeline Test ---
rag_system = FullRAGPipeline(database_source=RAW_KNOWLEDGE_BASE)

# Test Query that targets our custom hardware design document profiles
query = "What switching transistors are used in the plasma singing arc project, and how do you cool them?"
final_answer = rag_system.execute_rag_flow(query)

print("FINAL SYSTEM REPLICA ANSWER:")
print(final_answer)
```
