```python
import json
from typing import Dict, Any, List
```

```python
class MemGPTOperatingSystem:
    """
    Simulates a self-managing memory architecture. 
    The model controls its own context space using dedicated memory tools.
    """
    def __init__(self):
        # 1. Core Memory Tiers (Always visible to the LLM in its prompt prompt block)
        self.core_memory = {
            "user_name": "Unknown",
            "current_project": "None Listed",
            "hardware_stack": "Unknown Workstation"
        }
        # 2. Deep Archival Storage (Hidden from model until explicitly searched via tool)
        self.archival_storage: List[str] = [
            "Archive Log 2025: User hosted local game distribution system 'CampusGrid' on hostel intranet.",
            "Archive Log 2026: User submitted bandgap predictor model using XGBoost to PeaceOfCode hackathon."
        ]
        print("[MemGPT OS] Virtual Memory Tiers initialized. Core Memory mapped to RAM space.")

    def get_internal_system_prompt(self) -> str:
        """Compiles the visible context window state presented to the model's attention layers."""
        return f"""
======================================================================
MEMGPT VIRTUAL OPERATING SYSTEM SYSTEM CONTEXT
======================================================================
CORE MEMORY (VISIBLE RAM):
{json.dumps(self.core_memory, indent=2)}

AVAILABLE MEMORY MANAGEMENT TOOLS:
- core_memory_update[key, new_value]: Modifies a fact in your visible RAM scratchpad.
- archival_search[query]: Queries your deep hard-drive databases to pull up historical logs.

TASK: Process the user input. If they share new long-term personal traits, update your Core Memory immediately using 'core_memory_update'. If they ask about past projects, invoke 'archival_search'.
======================================================================
"""

    # --- Underlying Tool Execution Logic ---
    
    def core_memory_update(self, key: str, new_value: str) -> str:
        """Allows the LLM to actively rewrite its own visible fact sheet."""
        if key in self.core_memory:
            old_val = self.core_memory[key]
            self.core_memory[key] = new_value
            return f"SUCCESS: Core Memory key '{key}' changed from '{old_val}' to '{new_value}'."
        return f"ERROR: Key '{key}' does not exist in Core Memory schema."

    def archival_search(self, query: str) -> List[str]:
        """Allows the LLM to scan its hidden historical archives on demand."""
        print(f"\n[OS DISK READ] Scanning hard-drive archives for query concept: '{query}'...")
        matches = [line for line in self.archival_storage if query.lower() in line.lower()]
        return matches if matches else ["No matching archived records located."]

    # --- Simulated Cognitive Execution Steps ---

    def handle_user_turn(self, user_message: str):
        print(f"\nUser Says: \"{user_message}\"")
        
        # Compile what the model currently sees inside its context window
        visible_context = self.get_internal_system_prompt()
        
        # --- Simulating LLM Cognitive Reasoning & Tool Selection ---
        # Scenario 1: User explicitly introduces himself
        if "aryan" in user_message.lower():
            print("\n[LLM Reasoning] I detected new critical user profile characteristics. I need to commit this to my visible Core Memory block.")
            
            # The model triggers its self-directed memory writing tool
            tool_output = self.core_memory_update("user_name", "Aryan Yadav")
            print(f"[Tool Execution Output] » {tool_output}")
            
        # Scenario 2: User asks about historical archives
        elif "hackathon" in user_message.lower() or "past" in user_message.lower():
            print("\n[LLM Reasoning] The user is asking about historical project milestones. This data is not in my visible RAM. I must trigger an archival disk search.")
            
            # The model triggers its archive scanning tool
            search_results = self.archival_search("hackathon")
            print(f"[Tool Execution Output] » Retrieved Context: {search_results}")
```

```python
# --- Run the MemGPT OS Engine Simulation ---
memgpt_system = MemGPTOperatingSystem()

# Execution Step Turn 1: Introduce information (Triggers self-directed write tool)
memgpt_system.handle_user_turn("Hey, just editing my profile name. Update it to aryan yadav.")

# Execution Step Turn 2: Query historical context (Triggers self-directed read tool)
memgpt_system.handle_user_turn("Can you pull up what model architecture I used for my past hackathon project submission?")

# Print final updated state of Core Memory to verify it successfully updated itself
print("\n" + "="*60)
print("FINAL VISIBLE CORE MEMORY RAM STATE AFTER SELF-MANAGEMENT:")
print("="*60)
print(json.dumps(memgpt_system.core_memory, indent=2))
```
