```python
import time
from typing import List, Dict, Any
```

```python
# --- Step 1: Define the Two Structural Memory Databases ---

# Chronological event log stream (Time-stamped diary)
EPISODIC_DATABASE_LOGS: List[Dict[str, Any]] = [
    {
        "timestamp": 1711958400, # Simulated Epoch time for March 2026
        "event_summary": "User ran PeaceOfCode Hackathon pipeline test. XGBoost model trained on Matbench database returned a baseline target bandgap prediction error of 0.22 eV."
    },
    {
        "timestamp": 1712044800, # Next day
        "event_summary": "User modified material training dataset features, adding crystalline symmetry markers to the input vector array."
    }
]

# Timeless conceptual dictionary profile (Vector database simulation)
SEMANTIC_KNOWLEDGE_GRAPH: Dict[str, Any] = {
    "project_name": "Materials Informatics Pipeline",
    "target_model_architecture": "XGBoost Regressor Core",
    "evaluation_metric": "Mean Absolute Error (MAE)"
}
```

```python
# --- Step 2: The Multi-Engine Memory Controller Subsystem ---

class LongTermMemoryController:
    def __init__(self):
        print("[MEMORY CONTROLLER] Long-Term Episodic and Semantic storage channels online.")

    def query_semantic_memory(self, concept_key: str) -> str:
        """Performs a direct factual look-up for timeless configuration variables."""
        print(f"  [Semantic Search] Searching vector/knowledge index for concept: '{concept_key}'...")
        if concept_key in SEMANTIC_KNOWLEDGE_GRAPH:
            return f"FACTUAL RECORD: The current state of '{concept_key}' is defined as '{SEMANTIC_KNOWLEDGE_GRAPH[concept_key]}'."
        return "Concept structural entity not found in semantic knowledge mapping."

    def query_episodic_memory_timeline(self, keywords: List[str]) -> List[str]:
        """Scans the chronological timeline logs to reconstruct historical sequences."""
        print(f"  [Episodic Scan] Scanning time-stamped experience registries for historical keywords: {keywords}...")
        matching_episodes = []
        
        for episode in EPISODIC_DATABASE_LOGS:
            summary_lower = episode["event_summary"].lower()
            # If all target keywords match the historical log string, capture it
            if any(word.lower() in summary_lower for word in keywords):
                readable_time = time.strftime('%Y-%m-%d %H:%M:%S UTC', time.gmtime(episode["timestamp"]))
                matching_episodes.append(f"[{readable_time}] Event Record: {episode['event_summary']}")
                
        return matching_episodes
```

```python
# --- Step 3: Run the Cognitive Routing Integration ---

memory_manager = LongTermMemoryController()

# SCENARIO A: User asks a purely factual, timeless question
print("\n--- Execution Scenario A: Factual Request ---")
query_a = "What machine learning architecture is our materials informatics pipeline currently using?"
# The system maps this to a Semantic Lookup
fact_sheet_response = memory_manager.query_semantic_memory("target_model_architecture")
print(f"Result Delivered to Agent Reasoning Step:\n  » {fact_sheet_response}")


# SCENARIO B: User asks a chronological historical question tracking past progress
print("\n--- Execution Scenario B: Timeline Review ---")
query_b = "Review what happened during my hackathon dataset calibration tests over the past few days."
# The system maps this to an Episodic Scan
historical_timeline = memory_manager.query_episodic_memory_timeline(keywords=["Hackathon", "XGBoost", "Matbench"])

print("Chronological History Array Reconstructed and Sent to LLM:")
for rank, episode_line in enumerate(historical_timeline, start=1):
    print(f"  Turn {rank}: {episode_line}")
```
