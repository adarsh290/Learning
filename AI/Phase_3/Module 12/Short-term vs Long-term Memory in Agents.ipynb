```python
import json
from typing import List, Dict, Any
```

```python
# --- Step 1: Define the Non-Volatile Long-Term Memory Profile Store ---
# In production, this would map directly to a permanent database like PostgreSQL or a Vector DB.
PERSISTENT_LONG_TERM_STORE = {
    "user_identity_01": {
        "user_name": "Aryan Yadav",
        "academic_institution": "IIT Ropar",
        "primary_hardware_spec": "HP Omen, R7 CPU, RTX 4060, 16GB RAM"
    }
}
```

```python
# --- Step 2: The Dual-Track Memory Architecture Agent ---

class DualTrackMemoryAgent:
    def __init__(self, profile_id: str):
        self.profile_id = profile_id
        # Initialize an empty, fresh short-term memory track for this live chat session
        self.short_term_context_history: List[Dict[str, str]] = []
        print("[MEMORY ENG] New active session initialized. Short-term memory cleared.")

    def add_to_short_term(self, role: str, message_text: str):
        """Appends the direct chronological conversational line to the live loop track."""
        self.short_term_context_history.append({"role": role, "content": message_text})

    def recall_long_term_profile(self) -> Dict[str, Any]:
        """Fetches persistent cross-session facts matching the active user identification."""
        print(f"[MEMORY ENG] Querying persistent database for profile: '{self.profile_id}'...")
        return PERSISTENT_LONG_TERM_STORE.get(self.profile_id, {})

    def compile_augmented_runtime_prompt(self, current_user_query: str) -> str:
        """
        The Integration Layer.
        Merges long-term context records and active short-term history into one prompt.
        """
        # 1. Fetch long-term facts
        long_term_facts = self.recall_long_term_profile()
        
        # 2. Extract active short-term turns
        formatted_short_term = ""
        for turn in self.short_term_context_history:
            formatted_short_term += f"{turn['role'].upper()}: {turn['content']}\n"
            
        # 3. Construct the complete composite context envelope prompt
        composite_prompt = f"""
SYSTEM PERSONA: You are a deeply personalized code optimization intelligence agent.

CROSS-SESSION LONG-TERM MEMORY RECORDS:
- User Name: {long_term_facts.get('user_name', 'Unknown')}
- Base Institution: {long_term_facts.get('academic_institution', 'Generic')}
- Verified Workstation Assets: {long_term_facts.get('primary_hardware_spec', 'Standard Hardware')}

ACTIVE SESSION SHORT-TERM CHAT LOGS:
{formatted_short_term}
USER CURRENT QUERY: {current_user_query}

REASONED ARCHITECTURAL RESPONSE:"""
        
        return composite_prompt

    def simulate_generation(self, complete_prompt: str) -> str:
        """Simulates an LLM evaluating the dual-track prompt boundaries."""
        print("[LLM Core] Analyzing compound memory matrix data lines...")
        
        # Check if the prompt properly contains our injected long-term asset variables
        if "RTX 4060" in complete_prompt and "16GB RAM" in complete_prompt:
            return "Based on your HP Omen workstation specifications (RTX 4060, 16GB RAM), we should optimize your PyTorch model allocation script to use native CUDA execution blocks and cap the dataloader worker processes to 4 to prevent RAM saturation."
        return "System optimization model trace ready."
```

```python
# --- Step 3: Run the Multi-Turn Execution Scenario ---

# Scenario: The user opens a completely fresh chat terminal on day 40 of using the app
agent = DualTrackMemoryAgent(profile_id="user_identity_01")

# Turn 1: Immediate interaction containing NO hardware descriptors
user_input_1 = "I am initializing a heavy local deep learning training run in PyTorch. How should I scale my pipeline configuration to prevent system lockups?"
agent.add_to_short_term("user", user_input_1)

# Build the prompt. Note how the long-term memories are seamlessly injected 
# even though the user didn't mention their laptop in this session!
runtime_prompt = agent.compile_augmented_runtime_prompt(current_user_query=user_input_1)

# Get the tailored contextual generation
assistant_reply = agent.simulate_generation(runtime_prompt)
agent.add_to_short_term("assistant", assistant_reply)

print("\n" + "="*60)
print("TAILORED PERSONALIZED AI RESPONSE GENERATED:")
print("="*60)
print(assistant_reply)
```
