```python
import torch
import torch.nn.functional as F
import math
```

```python
# Set dimensions for a example
seq_len = 3
d_k = 4  # Dimension of keys/queries

# Simulating Q, K, V tensors (Batch size = 1)
# Shape: [batch_size, seq_len, d_k]
Q = torch.randn(1, seq_len, d_k)
K = torch.randn(1, seq_len, d_k)
V = torch.randn(1, seq_len, d_k)
```

```python
# Step 1 & 2: Calculate Raw Attention Scores and Scale them
# We need to multiply Q [1, 3, 4] by K transposed. 
# To transpose only the last two dimensions, we use K.transpose(-2, -1)
scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(d_k)
print("Scaled Scores Shape [batch, seq_len, seq_len]:", scores.shape)

# Step 3: Apply Softmax along the very last dimension to get weights
attention_weights = F.softmax(scores, dim=-1)
print("\nAttention Weights (Rows sum to 1):\n", attention_weights)

# Step 4: Multiply weights by V to get the final output context vectors
output = torch.matmul(attention_weights, V)
print("\nFinal Output Shape [batch, seq_len, d_v]:", output.shape)
```

```python
import torch
import torch.nn.functional as F
import math

def scaled_dot_product_attention(Q, K, V):
    # 1. Get the dimension of the keys (d_k) from the last dimension of K
    # Hint: K.shape[-1] gives you this value.
    d_k = K.shape[-1]
    # 2. Compute raw attention scores (Q multiplied by K transposed)
    # Remember to transpose ONLY the last two dimensions of K!
    raw_attention_scores = (torch.matmul(Q, K.transpose(-2,-1)))
    # 3. Scale the scores by dividing by the square root of d_k
    scaled_score = raw_attention_scores/math.sqrt(d_k)
    # 4. Apply softmax along the last dimension (dim=-1) to get weights
    attention_weights = F.softmax(scaled_score, dim=-1)
    # 5. Multiply weights by V to get the final output context vectors
    output = torch.matmul(attention_weights, V)
    return output, attention_weights

# --- Test Data ---
torch.manual_seed(42)
B, SeqLen, D = 1, 3, 4  # Batch=1, Tokens=3, Dimension=4

test_Q = torch.randn(B, SeqLen, D)
test_K = torch.randn(B, SeqLen, D)
test_V = torch.randn(B, SeqLen, D)

out, weights = scaled_dot_product_attention(test_Q, test_K, test_V)

print("Output shape:", out.shape)
print("Weights shape:", weights.shape)
print("\nWeights matrix:\n", weights)
```
