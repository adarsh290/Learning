```python
import torch
import torch.nn as nn
import math
```

```python
# Set seed for reproducibility
torch.manual_seed(42)
```

```python
# Configuration parameters
batch_size = 1       # Processing one sequence at a time
seq_len = 3          # Sentence length: e.g., "I love AI" (3 tokens)
embed_dim = 4        # Each word is represented by a vector of size 4
```

```python
# 1. Simulate input embeddings (X)
# Shape: [batch_size, seq_len, embed_dim]
X = torch.randn(batch_size, seq_len, embed_dim)
print("Input Shape (X):", X.shape)
```

```python
# 2. Define the linear projection layers (The W_Q, W_K, W_V matrices)
# These layers hold the weights that the model learns during training.
# They map our input embedding dimension to our target projection dimension.
# For simplicity, we keep the output size the same as embed_dim (4).
W_q = nn.Linear(embed_dim, embed_dim, bias=False)
W_k = nn.Linear(embed_dim, embed_dim, bias=False)
W_v = nn.Linear(embed_dim, embed_dim, bias=False)
```

```python
# 3. Project the input X to generate Queries, Keys, and Values
Q = W_q(X)  # Q = X * W_Q
K = W_k(X)  # K = X * W_K
V = W_v(X)  # V = X * W_V

print("Queries Shape (Q):", Q.shape)
print("Keys Shape (K):   ", K.shape)
print("Values Shape (V): ", V.shape)
```

# Exercise

```python
# A single token embedding vector (size 3)
x = torch.tensor([1.0, 2.0, -1.0])
```

```python
# Weight matrices (each of shape [hidden_dim, embed_dim] -> [2, 3])
# This means your output Q, K, V vectors should each have a size of 2.
W_Q = torch.tensor([[1.0, 0.0, 2.0],
                    [0.0, 1.0, -1.0]])

W_K = torch.tensor([[0.0, 1.0, 1.0],
                    [2.0, -1.0, 0.0]])

W_V = torch.tensor([[1.0, 1.0, 0.0],
                    [0.0, 0.0, 3.0]])
```

```python
q = x @ W_Q.T
k = x @ W_K.T
v = x @ W_V.T
```

```python
print("Query vector q:", q)
print("Key vector k:", k)
print("Value vector v:", v)
```
