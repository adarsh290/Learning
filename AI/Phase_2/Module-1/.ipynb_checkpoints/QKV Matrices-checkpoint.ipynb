```python
import torch
import torch.nn as nn
import math
```

```python
class QKVSetup(nn.Module):
    def __init__(self, d_model, head_dim):
        super().__init__()
        # We use linear layers to project the input into Q, K, and V
        self.w_q = nn.Linear(d_modle, head_dim, bias=False)
        self.w_k = nn.Linear(d_modle, head_dim, bias=False)
        self.w_v = nn.Linear(d_modle, head_dim, bias=False)
        
    def forward(self, x):
        q = self.w_q(x)
        k = self.w_k(x)
        v = self.w_v(x)
        return q, k, v
```

```python
def scaled_dot_product_attention(q, k, v, mask):
    scores = torch.matmul(q, k.transpose(-2, -1))
    d_k = q.size(-1)
    scores = scores / math.sqrt(d_k)
    if mask is not None:
        scores = scores.masked_fill(mask == 0, float('-inf'))
    attention_weights = torch.softmax(scores, dim=-1)
    output = torch.matmul(attention_weights, v)  # (batch, seq_len, d_k)

    return output, attention_weights
```
