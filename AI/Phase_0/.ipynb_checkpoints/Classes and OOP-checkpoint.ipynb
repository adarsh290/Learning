# Example

In AI production and PyTorch, models are structured as Object-Oriented Programming (OOP) classes. A class acts as a blueprint that bundles state (like layer weights and biases) and behavior (like the forward pass operation) into a single entity. This approach ensures state persistence and enables modular layer reuse.

```python
# A minimal blueprint representing a simple neural component
class Perceptron:
    # The __init__ constructor initializes the object's internal state
    def __init__(self, weight, bias):
        self.w = weight
        self.b = bias

    # An instance method defining how the object processes input data
    def forward(self, x):
        # Accesses the instance state to compute a linear transformation
        return (x * self.w) + self.b

# Instantiate the object with specific weight and bias state
node = Perceptron(weight=2.5, bias=1.0)

# Pass data through the instance
result = node.forward(x=4.0)
print(f"Prediction: {result}")  # Output: 11.0
```

# Exercise

Create a class called SingleLayerPerceptron that initializes with a weight vector (a standard Python list of numbers) and a single bias value. Implement a forward(x) method that takes an input vector (list of numbers of the same length), computes the dot product (w⋅x+b), and returns the result.

```python
class SingleLayerPerceptron:
    
    def __init__(self, weight, bias):
        self.w = weight
        self.b = bias

    def forward(self, x):
        dot_prod=0
        for a, b in zip(self.w, x):
            dot_prod += a*b
        return dot_prod + self.b
```

# Example

Inheritance allows a new class (subclass) to inherit methods and properties from an existing class (superclass). In AI engineering, this is the foundational architecture of PyTorch: every custom neural network model you build must inherit from torch.nn.Module, giving your model automatic access to optimization tracking, parameter management, and GPU capabilities.

```python
# Superclass: Represents a generic component with a state check
class BaseLayer:
    
    def __init__(self, name):
        self.layer_name = name

    def debug_info(self):
        return f"Layer: {self.layer_name}"

# Subclass: Inherits from BaseLayer
class ActivationLayer(BaseLayer):
    # If __init__ is omitted, it automatically uses the superclass __init__
    def set_threshold(self, threshold):
        self.threshold = threshold

    def relu(self, x):
        return max(0.0, x)
```

```python
# Instantiate the subclass
layer = ActivationLayer("ReLU_1")
# Accessing a method from the superclass
print(layer.debug_info())  # Output: Layer: ReLU_1
# Accessing a method from the subclass
print(layer.relu(-4.5))    # Output: 0.0
```

# Exercise

Create a superclass named Parameters that initializes with a learning rate (lr). Then, create a subclass named SGD (Stochastic Gradient Descent) that inherits from Parameters. The SGD class should implement an update(weight, gradient) method that calculates and returns the updated weight using the standard formula:
                                            updated_weight = weight − (lr×gradient)

```python
class Parameters:
    def __init__(self, learning_rate):
        self.lr = learning_rate

class SGD(Parameters):
    def update(self, weight, gradient):
        return weight - (self.lr*gradient)
```

# Example

When a subclass defines its own __init__ method, it overrides the superclass's constructor entirely. To leverage the superclass setup while adding subclass-specific state, you must explicitly call super().__init__(). This pattern is ubiquitous in PyTorch, where you initialize the base nn.Module before defining your custom network layers.

```python
class ParentModel:
    def __init__(self, device):
        self.device = device
        print(f"Model pinned to device: {self.device}")

class CustomTransformer(ParentModel):
    def __init__(self, device, num_layers):
        # Explicitly invoke the parent constructor to set 'device'
        super().__init__(device) 
        # Set subclass-specific parameters
        self.num_layers = num_layers
```

```python
# Instantiating triggers both configuration steps
gpt = CustomTransformer(device="cuda", num_layers=12)
# Output:
# Model pinned to device: cuda
```

# Example

PyTorch models do not expose a .predict() method like scikit-learn does. Instead, they implement a special method called __call__. In Python, defining __call__ allows an instance of a class to be executed exactly like a function. In AI pipelines, this maps directly to running a model's forward pass.

```python
class SoftmaxActivation:
    def __init__(self, temperature):
        self.temp = temperature

    # Defining __call__ makes the object instance callable
    def __call__(self, logit):
        # A mock scaled activation step
        return logit / self.temp
```

```python
# Instantiate the object
activate = SoftmaxActivation(temperature=2.0)

# CRITICAL DISTINCTION: We do NOT call activate.__call__(10.0)
# We treat the object instance itself as a function:
output = activate(10.0) 

print(f"Activated Output: {output}")  # Output: 5.0
```

# Exercise

Create a class called DenseLayer. It should initialize with an integer parameter units. Implement a __call__ method that accepts a single list of numbers representing inputs. Inside __call__, multiply every input element by the total number of units and return the modified list.

```python
class DenseLayer:
    def __init__(self, units):
        self.units = units

    def __call__(self, inputs):
        return [x * self.units for x in inputs]
```

# FINAL EXERCISE

Objective:
Build a complete LinearRegression model class from scratch using pure Python OOP mechanics to solidify your state management and forward pass syntax skills before we add array math libraries.

Requirements:
1. Create a class called LinearRegression.
2. The __init__ constructor must accept and store two parameters: weights (a list of numbers) and bias (a single float/int).
3. Implement a predict(x) method that accepts an input vector x (a list of numbers matching the length of weights).
4. Inside predict(x), compute the standard linear equation:{ y=(w⋅x)+b }.
   This requires calculating the dot product of the weights and the inputs, then adding the bias.
5. Ensure your method returns the calculated final numeric value.

Verification Task:
Once written, instantiate your class with weights=[2.0, -1.5, 0.5] and bias=1.0. Call the predict() method passing the input vector x=[4.0, 2.0, 1.0].

```python
class LinearRegression:
    def __init__(self, weight, bias):
        self.w = weight
        self.b = bias
    def predict(self, x):
        dot_prod = 0
        for a, b in zip(self.w, x):
            dot_prod += a*b
        return dot_prod + self.b
```

```python
weights=[2.0, -1.5, 0.5]
bias=1.0
x=[4.0, 2.0, 1.0]
model = LinearRegression(weights, bias)
output = model.predict(x)
print(output)
```
