# Example

Data cleaning and log filtering pipelines process raw files using Python functions. To build highly scalable architectures, you must master variable unpacking using *args (positional arguments collected as a tuple) and kwargs (keyword arguments collected as a dictionary). This allows wrapper functions to dynamically pass settings down to model components.

```python
# A wrapper that logs meta-parameters before executing an operation
def hyperparameter_logger(model_name, *args, **kwargs):
    # *args collects positional variables into a tuple
    if args:
        print(f"Positional settings (tuple): {args}")
    # **kwargs collects key-value pairs into a dictionary
    if kwargs:
        print(f"Keyword settings (dict): {kwargs}")

# Executing the function with mixed configurations
hyperparameter_logger("ResNet-50", 224, 224, learning_rate=0.001, optimizer="Adam")
```

# Example

List comprehensions provide an optimized syntax to create lists out of existing sequences. In text processing pipelines (like tokenization and vocabulary generation), they run significantly faster than standard for loops because the looping operation is optimized under the hood in Python's runtime environment.

```python
# Raw token tokens containing trailing newline tags and punctuation noise
raw_tokens = ["  \nattention  ", "TRANSFORMER", "  embedding\t", "the"]

# Standard loop approach vs List Comprehension
# Format: [expression for item in iterable if condition]
cleaned_tokens1 = [tok.strip().lower() for tok in raw_tokens if len(tok.strip()) > 3]
cleaned_tokens2 = [tok.lower() for tok in raw_tokens if len(tok.strip()) > 3]
cleaned_tokens3 = [tok.strip() for tok in raw_tokens if len(tok.strip()) > 3]
print(cleaned_tokens1)
print(cleaned_tokens2)
print(cleaned_tokens3)
# ("the" was dropped because its stripped length is not > 3)
```

# Example

Anonymous one-liner functions in Python are called lambda functions. They follow the syntax lambda arguments: expression. In data science pipelines, they are used for instant, throwaway transformations within high-order functions like map() (which applies an operation over an iterable) or sorting routines, avoiding the need to declare a full named function blocks.

```python
# A list of numeric probabilities
probabilities = [0.12, 0.85, 0.43, 0.91]

# map() applies the lambda scale formula across the iterable elements
# Syntax: lambda input_val: output_calculation
scaled_probs = list(map(lambda p: p * 100, probabilities))

print(scaled_probs)  

# Sorting complex tuple records by specific index targets
dataset_records = [("batch_C", 14), ("batch_A", 92), ("batch_B", 5)]
# Use lambda to sort by the integer score (index 1 of each tuple) instead of the string
dataset_records.sort(key=lambda record: record[1])

print(dataset_records)  # Output: [('batch_B', 5), ('batch_C', 14), ('batch_A', 92)]
```

# Exercise

You have a list of prediction threshold tuples:
thresholds = [(0.5, "Class_B"), (0.2, "Class_C"), (0.9, "Class_A")]

Use the built-in sorted() function or the .sort() list method along with a custom lambda function to sort these tuples in descending order based on their float value (index 0). Print the sorted list.

```python
thresholds = [(0.5, "Class_B"), (0.2, "Class_C"), (0.9, "Class_A")]
sorted_thresholds = sorted(thresholds, key=lambda x: x[0], reverse=True)

print(sorted_thresholds)
```

# Example

Data engineering often requires matching parallel data sequences (e.g., matching a list of text strings with a separate list of predicted probability labels). The zip() function binds multiple iterables together, element-by-element, returning an iterator of tuples. Combining zip() with dict() instantly maps parallel data series into an accessible lookup table.

```python
# Two separate, parallel data tracks of equal length
token_strings = ["attention", "weights", "bias"]
dimension_sizes = [512, 1024, 64]

# zip() pairs index 0 with index 0, index 1 with index 1, etc.
zipped_pairs = list(zip(token_strings, dimension_sizes))
print(zipped_pairs)

# Passing a zipped sequence directly to dict() builds a quick lookup map
config_map = dict(zip(token_strings, dimension_sizes))
print(config_map)
# Output: {'attention': 512, 'weights': 1024, 'bias': 64}
print(config_map["bias"])  # Output: 64
```

keys = ["learning_rate", "batch_size", "epochs"]
values = [0.001, 32, 10]

# FINAL EXERCISE

Objective:
Build a text processing pipeline function called clean_text. In natural language processing (NLP), raw input text corpora are always filled with noisy formatting, whitespace discrepancies, and short garbage tokens. Cleaning this data efficiently is the first step before generating token embeddings.

Requirements:
1. Write a function called clean_text(texts) that takes a single positional argument: a list of strings.
2. Inside the function, use a single-line list comprehension to clean and filter the strings.
3. For every string in the input list, your list comprehension must:
4. Strip away any leading and trailing whitespace characters.
5. Convert all characters to lowercase.
6. Filter out and completely remove any string that is shorter than 3 characters after it has been stripped.
7. The function must return the newly generated list of clean strings.

Verification Task:
Once written, test your function by passing this dirty text dataset:
dirty_corpus = ["  \nTransformer  ", "  ai ", "RAG", "  \t  ", "llm", " o "]

Print the final output of your function to verify that the formatting noise and short strings are completely eliminated.

```python
def clean_text(texts):
    return [text.strip().lower() for text in texts if len(text.strip()) >= 3]
```

```python
dirty_corpus = ["  \nTransformer  ", "  ai ", "RAG", "  \t  ", "llm", " o "]

clean_text = clean_text(dirty_corpus)
print(clean_text)
```
