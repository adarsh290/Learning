```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
class SimpleNetwork(nn.Module):
    def __init__(self):
        
        # IMP: We should first call the parent class constructor.
        super().__init__()

        # Define the layers. 
        # nn.Linear represents a fully connected layer. Mathematically: y = xA^T + b
        # It automatically creates and tracks the weight matrix and bias vector for us.
        # Let's say we have an input of 10 features, and we want 64 hidden nodes.
        self.layer1 = nn.Linear(10,64)

        # The input of this next layer MUST match the output of the previous layer (64).
        # We want our final output to be 2 numbers.(all the numbers are random)
        self.layer2 = nn.Linear(64, 2)

        
    def forward(self, x):
        
        # 1. Pass input x through layer 1
        x = self.layer1(x)

        # 2. Apply an activation function (ReLU). 
        # This introduces non-linearity. Without this, stacking layers is mathematically useless 
        # because the whole network collapses into a single linear transformation.
        x = F.relu(x)

        # 3. Pass through layer 2 to get the final prediction
        x = self.layer2(x)
        
        return x
```

```python
# 1. Setup: Assume we have the SimpleNetwork from the last checkpoint
model = SimpleNetwork()
```

```python
# 2. Define the Loss Function (Criterion) and Optimizer
# MSE measures how far our predictions are from the true answers.
criterion = nn.MSELoss() 
# SGD will update the weights. lr=0.01 is the Learning Rate (step size).
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# Dummy Data: 1 sample, 10 features. True answer we want the model to output is [1.0, 0.0]
inputs = torch.randn(1, 10)
targets = torch.tensor([[1.0, 0.0]])
```

```python
# 3. The Training Loop (Let's run 100 "epochs" or iterations)
epochs = 100
for epoch in range(epochs):
    
    # STEP 1: Clear old gradients. NEVER FORGET THIS.
    optimizer.zero_grad()
    
    # STEP 2: Forward Pass. Get the model's current garbage prediction.
    outputs = model(inputs)
    
    # STEP 3: Calculate the Loss. Compare prediction vs reality.
    loss = criterion(outputs, targets)
    
    # STEP 4: Backward Pass. Calculate the gradients (the required corrections).
    loss.backward()
    
    # STEP 5: Optimizer Step. Apply the corrections to the weights.
    optimizer.step()
    
    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}/100, Loss: {loss.item():.4f}")
```
