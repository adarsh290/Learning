# nn.Module
We need a system to cleanly organize these weights and the mathematical operations that connect them. nn.Module is PyTorch's solution.

To build a network, you strictly need to define two things inside your class:

    __init__(self): The constructor. This is where you declare your layers (the "physical" parts of your network).

    forward(self, x): The execution pipeline. This is where you define exactly how your input tensor x flows through the layers you declared in the constructor.


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class SimpleNetwork(nn.Module):
    def __init__(self):
        
        # IMP: We should first call the parent class constructor.
        super().__init__()

        # Define the layers. 
        # nn.Linear represents a fully connected layer. Mathematically: y = xA^T + b
        # It automatically creates and tracks the weight matrix and bias vector for us.
        # Let's say we have an input of 10 features, and we want 64 hidden nodes.
        self.layer1 = nn.Linear(10,64)

        # The input of this next layer MUST match the output of the previous layer (64).
        # We want our final output to be 2 numbers.(all the numbers are random)
        self.layer2 = nn.Linear(64, 2)

        
    def forward(self, x):
        
        # 1. Pass input x through layer 1
        x = self.layer1(x)

        # 2. Apply an activation function (ReLU). 
        # This introduces non-linearity. Without this, stacking layers is mathematically useless 
        # because the whole network collapses into a single linear transformation.
        x = F.relu(x)

        # 3. Pass through layer 2 to get the final prediction
        x = self.layer2(x)
        
        return x
```

```python
# Instantiate the network
model = SimpleNetwork()
```

```python
# Create a dummy input tensor: let's pretend it's 1 row of data with 10 features.
dummy_input = torch.randn(1, 10)
```

```python
# Pass the input through the model to get a prediction.
# Notice we do NOT call model.forward(dummy_input). We just call the model object directly like a function.
output = model(dummy_input)
print(output)
```

# Excersise 1.2

```python
class MyFirstNetwork(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(5, 16)
        self.layer2 = nn.Linear(16, 8)
        self.layer3 = nn.Linear(8, 1)
        
    def forward(self, x):
        x = self.layer1(x)
        x = F.relu(x)
        x = self.layer2(x)
        x = F.relu(x)
        x = self.layer3(x)
        return x
mymodel = MyFirstNetwork() 
mydummy_input = torch.rand(1, 5)
mymodel(mydummy_input)
```
