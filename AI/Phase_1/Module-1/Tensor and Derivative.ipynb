```python
import torch
```

# 1. Create a tensor.
Think of this as defining an initial input variable.

We set requires_grad=True to tell PyTorch, "Track every mathematical operation I do to 'x'."

```python
x = torch.tensor([2.0], requires_grad=True)
```

# 2. Perform an operation.
 Mathematically: y = x^2 + 3x

```python
y = x**2 + 3*x
```

# 3. Calculate the gradients automatically.
This traverses the graph backwards. In calculus terms, it computes dy/dx.

```python
y.backward()
```

# 4. Check the result.
The analytical derivative of x^2 + 3x is 2x + 3 and at x = 2.0, this is 2(2) + 3 = 7.

Let's see what Autograd calculated and stored inside x.grad:

```python
print(x.grad)
```
