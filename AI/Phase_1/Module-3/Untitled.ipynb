```python
import torch
import torch.nn as nn
from torchvision import models
```

```python
# Goal: Train a model to classify 5 types of flowers
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
```

```python
for param in model.parameters():
    param.requires_grad = False # We want it to learn fast
```

```python
num_features = model.fc.in_features
```

```python
model.fc = nn.Linear(num_features, 5) # Set up for our 5 flower classes
```
