```python
import torch
import torch.nn as nn
```

```python
# 1. Create a dummy "image" 
# Format: [Batch Size, Channels (RGB), Height, Width]
# Let's make one 10x10 grayscale image (1 channel)
image = torch.randn(1, 1, 10, 10)
```

```python
# 2. Define a Convolutional Layer
# in_channels=1 (grayscale), out_channels=16 (we want to find 16 different patterns)
# kernel_size=3 (our 3x3 magnifying glass)
conv_layer = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3)
```

```python
# 3. Define a Pooling Layer
# kernel_size=2 (looks at 2x2 blocks and picks the max value)
pool_layer = nn.MaxPool2d(kernel_size=2)
```

```python
# 4. Pass the image through
features = conv_layer(image)
downsampled = pool_layer(features)
```

```python
print(f"Original shape: {image.shape}")      # [1, 1, 10, 10]
print(f"After Conv shape: {features.shape}") # [1, 16, 8, 8] -> It got smaller because of the filter size
print(f"After Pool shape: {downsampled.shape}") # [1, 16, 4, 4] -> It halved because of the 2x2 pool
```

# Exercise

```python
# 1. The 4x4 Image (Batch=1, Channels=1, H=4, W=4)
image = torch.tensor([[[
    [1., 0., 1., 0.],
    [0., 2., 0., 2.],
    [1., 0., 1., 0.],
    [0., 2., 0., 2.]
]]])
```

```python
# 2. Define the Conv Layer
# kernel_size=3, stride=1, padding=0 (default)
# We set bias=False so it doesn't add extra numbers to our sum
conv_exc = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, bias=False)
```

```python
# 3. Create our "Plus" Filter (1, 1, 3, 3)
plus_filter = torch.tensor([[[
    [0., 1., 0.],
    [1., 1., 1.],
    [0., 1., 0.]
]]])
```

```python
# 4. FORCE the layer to use our filter
# .data is used here to overwrite the internal weights
conv_exc.weight.data = plus_filter
```

```python
# 5. Define the Pooling Layer
# 2x2 window, 2x2 stride (non-overlapping)
pool = nn.MaxPool2d(kernel_size=2)
```

```python
# EXECUTION
with torch.no_grad(): # We aren't training, just calculating
    conv_result = conv_exc(image)
    final_result = pool(conv_result)

print("--- Convolution Result (2x2) ---")
print(conv_result[0, 0]) 

print("\n--- Final MaxPool Result (1x1) ---")
print(final_result[0, 0])
```
