```python
import torch
import torch.nn as nn
```

```python
class MyResBlock(nn.Module):
    def __init__(self, channels):
        super(MyResBlock, self).__init__()
        # Layer 1
        self.conv1 = nn.Conv2d(channels, channels, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(channels)
        self.relu = nn.ReLU()
        
        # Layer 2
        self.conv2 = nn.Conv2d(channels, channels, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(channels)

    def forward(self, x):
        # YOUR CODE HERE
        # 1. Create a copy of the input for the shortcut
        shortcut = x
        # 2. Process x through conv1 -> bn1 -> relu
        residual = self.conv1(x)
        residual = self.bn1(residual)
        residual = self.relu(residual)
        # 3. Process that result through conv2 -> bn2
        residual = self.conv2(residual)
        residual = self.bn2(residual)
        # 4. Add the shortcut back
        out = residual + shortcut
        # 5. Apply the final relu and return
        return self.relu(out)
```
