```python
import torch
import torch.nn as nn
```

```python
class ResidualBlock(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        # A simple block with two convolutions
        self.conv_path = nn.Sequential(
            nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(in_channels, in_channels, kernel_size=3, padding=1)
        )
        self.relu = nn.ReLU()

    def forward(self, x):
        # 1. Save the original input (the "Identity" or "Shortcut")
        shortcut = x
        # 2. Pass the input through the math layers
        residual = self.conv_path(x)
        # 3. ADD them together! (This is the magic part)
        out = residual + shortcut
        # 4. Final activation
        return self.relu(out)
```

```python
# Try it out
block = ResidualBlock(in_channels=64)
input_tensor = torch.randn(1, 64, 32, 32)
output = block(input_tensor)
print(output.shape) # Should be the same: [1, 64, 32, 32]
```
