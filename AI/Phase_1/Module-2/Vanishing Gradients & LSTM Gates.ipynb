```python
import torch
import torch.nn as nn
```

```python
# Parameters: input_size (features per word), hidden_size (memory units)
input_size = 10 
hidden_size = 20
num_layers = 1
```

```python
# Define the LSTM
lstm = nn.LSTM(input_size, hidden_size, num_layers)

# Create dummy input: (Sequence_Length, Batch_Size, Input_Size)
# Example: A sentence of 5 words, 3 sentences in a batch, each word has 10 features
input_data = torch.randn(5, 3, 10)

# Initialize Hidden state (h0) and Cell state (c0) with zeros
h0 = torch.zeros(1, 3, 20)
c0 = torch.zeros(1, 3, 20)
```

```python
# Forward pass
output, (hn, cn) = lstm(input_data, (h0, c0))

print(f"Output shape (all hidden states): {output.shape}") 
print(f"Final Hidden State (hn) shape: {hn.shape}")
print(f"Final Cell State (cn) shape: {cn.shape}")
```
