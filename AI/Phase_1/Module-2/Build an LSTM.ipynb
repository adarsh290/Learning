```python
import torch
import torch.nn as nn
```

```python
class SentimentLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SentimentLSTM, self).__init__()
        # 1. The LSTM "Memory" machine
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        # 2. The "Classifier" machine (converts memory to prediction)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        
        # x shape: (batch_size, seq_len, input_size)
        # lstm_out: all hidden states
        # (hn, cn): the final hidden and cell states
        lstm_out, (hn, cn) = self.lstm(x)
        # We only care about the memory AFTER the last word was read
        # hn shape: (num_layers, batch_size, hidden_size)
        # We take hn[0] to get the last layer's hidden state
        last_hidden_state = hn[0]
        prediction = self.fc(last_hidden_state)
        return prediction
```

```python
# Initialize: 10 input features, 20 memory units, 1 output (Positive/Negative)
model = SentimentLSTM(10, 20, 1)
print(model)
```

# Exercise
The Setup:

Input Size: 50 (Each word is represented by 50 numbers).

Hidden Size: 64 (The LSTM's internal memory).

Output Size: 2 (We want 2 scores: one for "Spam", one for "Not Spam").

```python
import torch
import torch.nn as nn

class TweetLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(TweetLSTM, self).__init__()
        
        # 1. Define the LSTM layer (use batch_first=True)
        self.lstm = nn.LSTM(50, 64, batch_first=True)
        
        # 2. Define the Linear layer (to map hidden_size to output_size)
        self.fc = nn.Linear(64, 2)
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_size)
        
        # 3. Pass input through LSTM
        # Hint: remember it returns (output, (hn, cn))
        lstm_out, (hn, cn) = self.lstm(x)
        
        # 4. Grab the final hidden state (hn) 
        # Since num_layers=1, use hn[0]
        last_hidden = hn[0]
        
        # 5. Pass that hidden state through the linear layer
        prediction = self.fc(last_hidden)
        
        return prediction

# Test your model
model = TweetLSTM(50, 64, 2)
dummy_input = torch.randn(32, 10, 50) # (Batch of 32, 10 words, 50 features)
result = model(dummy_input)

print(f"Output shape: {result.shape}") # Should be (32, 2)
```
