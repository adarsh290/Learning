```python
def plot_prediction(train_data, train_labels, test_data, test_labels, predictions):
    """
    Plots training data, test data, and compares predictions against ground truth.
    """
    plt.figure(figsize=(10, 7))
    
    # Helper to convert tensors to numpy if necessary
    def to_numpy(x):
        return x.cpu().numpy() if hasattr(x, 'cpu') else x

    # Plot training data in blue
    plt.scatter(to_numpy(train_data), to_numpy(train_labels), c="b", s=4, label="Training data")
    
    # Plot testing data in green
    plt.scatter(to_numpy(test_data), to_numpy(test_labels), c="g", s=4, label="Testing data")
    
    # Plot predictions in red
    if predictions is not None:
        plt.scatter(to_numpy(test_data), to_numpy(predictions), c="r", s=4, label="Predictions")
    
    # Formatting
    plt.title("Model Predictions vs Ground Truth")
    plt.xlabel("Input Feature")
    plt.ylabel("Output Label")
    plt.legend()
    plt.grid(True, alpha=0.3) # Adds a light grid for readability
    plt.show()
```

```python
plot_predction()
```
