```python
import socket
import time
import random
import math

host = '0.0.0.0'
port = 50505

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    s.bind((host, port))
    s.listen()
    print(f"Server (Sabotaged) listening on {host}:{port}...")
    
    conn, addr = s.accept()
    with conn:
        print(f"Connected by {addr}")
        iteration = 0
        while True:
            data = conn.recv(1024)
            if not data:
                break
            
            # --- SABOTAGE SECTION ---
            # We create a fake pattern for your ARIMA model to find later
            # 1. Trend: It gets slower over time (0.001 * iteration)
            # 2. Seasonality: A sine wave delay
            # 3. Noise: Random jitter
            base_delay = 0.1
            trend = iteration * 0.002
            seasonality = 0.05 * math.sin(iteration * 0.5)
            noise = random.uniform(0, 0.02)
            
            total_delay = base_delay + trend + seasonality + noise
            
            # Sleep to simulate network lag
            if total_delay > 0:
                time.sleep(total_delay)
            # ------------------------

            conn.sendall(data)
            iteration += 1
```

```python
import socket
import time
import matplotlib.pyplot as plt

host = '127.0.0.1' # Or your LAN IP if using two computers
port = 50505

latencies = []

print("Starting 60-second ping test...")

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.connect((host, port))
    
    for i in range(60):
        # 1. Start Stopwatch
        start_time = time.time()
        
        # 2. Send Data
        s.sendall(b"Ping")
        
        # 3. Wait for Reply
        data = s.recv(1024)
        
        # 4. Stop Stopwatch
        end_time = time.time()
        
        # 5. Calculate Latency (Round Trip Time)
        rtt = end_time - start_time
        latencies.append(rtt)
        
        print(f"Ping {i+1}: {rtt:.4f} seconds")

print("Data Collection Complete.")

# --- VISUALIZATION ---
# Show the data we just captured
plt.figure(figsize=(10, 5))
plt.plot(latencies, marker='o')
plt.title("Measured Network Latency (The Data for ARIMA)")
plt.xlabel("Ping Sequence")
plt.ylabel("Latency (Seconds)")
plt.show()

# Print the raw list so you can copy it if needed
print("RAW DATA for ARIMA:")
print(latencies)
```
